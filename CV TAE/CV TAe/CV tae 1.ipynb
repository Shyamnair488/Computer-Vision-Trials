# Find contours

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Use this for displaying images in Colab
```

```python
# Load the image
image = cv2.imread('/content/drive/MyDrive/CV TAe/2.png')
```

```python
# Display the original image
print("Original Image")
cv2_imshow(image)
```

```python
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Find Canny edges
edged = cv2.Canny(gray, 30, 200)
```

```python
# Display the image with Canny edges
print("Canny Edges")
cv2_imshow(edged)
```

```python
# Finding Contours
contours, hierarchy = cv2.findContours(edged.copy(),
                                       cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

print("Number of Contours found = " + str(len(contours)))
```

```python
# Draw all contours
cv2.drawContours(image, contours, -1, (0, 255, 0), 3)
# Display the image with contours
print("Image with Contours")
cv2_imshow(image)
```

# Text Recognition using OpenCV and Tesseract (OCR)

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
# Function to detect text in contours
def contours_text(orig, img, contours):
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        # Drawing a rectangle on copied image
        rect = cv2.rectangle(orig, (x, y), (x + w, y + h), (0, 255, 255), 2)

        # Display the image using matplotlib
        plt.imshow(cv2.cvtColor(rect, cv2.COLOR_BGR2RGB))
        plt.title('Detected Contour')
        plt.axis('off')
        plt.show()

        # Cropping the text block for OCR
        cropped = orig[y:y + h, x:x + w]

        # Apply OCR on the cropped image
        config = ('-l eng --oem 1 --psm 3')
        text = pytesseract.image_to_string(cropped, config=config)
        print(text)
```

```python
# Text recognition
def text_recognition(image_path):
    # Read image
    im = cv2.imread(image_path)

    # Apply OCR on the entire image
    config = ('-l eng --oem 1 --psm 3')
    text = pytesseract.image_to_string(im, config=config)

    # Print text and split lines
    lines = text.split('\n')
    for line in lines:
        print(line)
```

```python
image_path = '/content/drive/MyDrive/CV TAe/images.png'
```

```python
# Display the original image
# Load the image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct color display
plt.title('Original Image')
plt.imshow(img_rgb)
plt.axis('off')  # Hide axis
plt.show()
```

```python
text_recognition(image_path)
```
